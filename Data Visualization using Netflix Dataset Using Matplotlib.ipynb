## Data Visualization Using Netflix Dataset Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Reading in the dataset

```python
df = pd.read_csv('/Users/moiznoorali/Documents/DataSets/Data Analytics Projects Kaggle/Netflix Dataset/netflix_titles.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

#### Summing all the null or 'NaN' values in dataset

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.duplicated().sum()
```

#### Displaying the Movie Counts by Country

```python
country_counts = df["country"].value_counts()
country_counts.head(10)
```

#### Top 10 Countries Based On Movie Counts

```python
df["country"].value_counts().head(10).plot(kind = "barh", rot = 0)
plt.title("Top 10 Movies Consumed by Country")
```

#### Displaying Movies vs. TV Shows Pie Chart

```python
show_counts = df["type"].value_counts()
show_counts.head()
```

```python
df["type"].value_counts().head(10).plot(kind = "pie", autopct = '%1.1f%%')
plt.title("Pie Chart Grouping Movies vs. TV Shows")
```

#### Based on Listed In

```python
listed_in = df["listed_in"].value_counts()
listed_in.head()
```

```python
df["listed_in"].value_counts().head().plot(kind = "pie", autopct = '%1.1f%%')
plt.title("Pie Chart Grouping Genres")
```

#### Seeing Films Being Released Over Time

```python
df['release_year'].plot(kind = "hist", bins = 50) 
plt.title("Distribution per year") 
plt.xlabel("Release year")
plt.ylabel("Count")
plt.show()
```

#### Movies / TV Shows Based on Duration

```python
duration = df["duration"].value_counts()
duration.head()
```

```python
duration.head(20).plot(kind = "barh", rot = 0) 
plt.title("Distribution per year") 
plt.xlabel("Release year")
plt.ylabel("Count")
plt.show()
```

#### Visualization Based on Rating Count

```python
rating = df["rating"].value_counts()
rating.head()
```

```python
df["rating"].value_counts().head().plot(kind = "pie", autopct = '%1.1f%%')
plt.title("Pie Chart Based On Show Ratings")
```

```python
director = df["director"].value_counts()
director.head(10)
```

```python
director.head(20).plot(kind = "barh", rot = 0) 
plt.title("To 10 Directors Count") 
plt.xlabel("Movies Directed")
plt.ylabel("Director Names")
plt.show()
```
